```python
#converting column names to lowercase and replacing spacess with '_'
df.columns = df.columns.str.replace(" ", "_").str.lower()
df.head()
```

```python
#standardizing the gender column
df['gender'] = df['gender'].str.strip().str.lower().str.capitalize()
```

```python
#one-hot encoding the gender for convinience
df = pd.get_dummies(df, columns=['gender'], prefix='is', prefix_sep='_')
df.head()
```

```python
#conerting booleans into int
df=df.astype(int)
```
